```python
from scipy.stats import norm

# Parámetros de la distribución normal
total_estudiantes = 500
media = 70        # kg
desviacion = 4    # kg

# Probabilidad de que un estudiante pese más de 68 y menos de 72 kg
# P(68 < X < 72)
probabilidad = norm.cdf(72, media, desviacion) - norm.cdf(68, media, desviacion)

print(f"Probabilidad de estudiantes que pesan más de 68 y menos de 72 kg: {probabilidad:.4f}")
print(f"En porcentaje: {probabilidad * 100:.2f}%")

# Cuántos estudiantes pesan más de 68 y menos de 72 kg
estudiantes = round(probabilidad * total_estudiantes)

print(f"\nDe {total_estudiantes} estudiantes, aproximadamente {estudiantes} pesan entre 68 y 72 kg.")
```
