### **Programa 4: distribucion binomial**



```python
from scipy.stats import binom

# Parámetros de la distribución binomial
n = 10
p = 0.10
x = 2

# Calcular la probabilidad de la variable aleatoria X = 2
probabilidad_x2 = binom.pmf(x, n, p)

print("La probabilidad de que la variable aleatoria X sea igual a 2 es:", probabilidad_x2)
```
