### **Programa 1: distribucion normal**



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

medi=70
desv_std=4
num_datos=500
datos=np.random.normal(medi, desv_std, num_datos)
datos[50]
x=np.linspace(medi - 3 * desv_std, medi + 3 * desv_std, num_datos)
pdf=1/(desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x-medi)**2/(2 * desv_std**2))
print ("Imprimir los primeros 50 valores de x", x[50], ", y los primeros 50 pdf", pdf[0:50])
plt.figure(figsize=(6,4))
plt.hist(datos,bins=15, alpha=0.4, color="b")
plt.title("Distribución Normal")
plt.xlabel("Peso en kg de los estudiantes")
plt.grid(True)
plt.show()
plt.figure(figsize=(6,4))
# Corrected: Removed 'pdf' from plt.hist arguments and changed 'media' to 'medi'
plt.hist(x, label="Normal:(media:{}; ds:{})". format(np.round(medi, 2), np.round(desv_std,2)))
plt.title("Funcion de densidad de probablidad distribucion normal")
plt.xlabel("Peso en kg de los estudiantes")
plt.ylabel("Densidad de probablidad")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,4))
plt.hist(datos,bins=15, alpha=0.4,density=True, color="b")
plt.plot(x, pdf, "k", linewidth=2, )

plt.title("Distribucion normal")
plt.xlabel("Peso en kg de los estudiantes")
plt.grid(True)
plt.show()
```
