### **Programa 1: ejemplo de distribucion de la probabilidad**



```python
import matplotlib.pyplot as plt
x = [5,6,7,8,9,10]
p=[0.5,0.10,0.20,0.30,0.25,0.10]
print("Suma de probabilidades:", sum(p))
plt.stem(x,p)
plt.title("Distribución de probabilidad")
plt.xlabel("Calificacion")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```
