```python
import math as mt
from itertools import combinations
from itertools import combinations_with_replacement
import numpy as np
nombres=["Oscar", "paco","Paty", "Laura","Ruben", "lUis", "Lucy", "Alberto","juan","daniel","Danny"]
n=len(nombres)
r=3
Cn=mt.factorial(n)/(mt.factorial(r)*mt.factorial(n-r))
print("Existen ", Cn , "posiles combinaciones sin repeticion del total de ", n, "caracteres en grupos de ", r, "en ", r)
print("Aqui las comnbinaciones")
combinaciones_nombres= list(combinations(nombres,r))
for combinacion in combinaciones_nombres:
    print(combinacion)
```
