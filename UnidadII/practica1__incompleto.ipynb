```python
import math as mt
from itertools import combinations

nombres = ["Oscar", "Paco", "Laura", "Ruben", "Luis", "Lucy", "Alberto", "Juan","Danny"]
n=len(nombres)
r=2 #¿como agrupar?

Cn= mt.factorial(n)/(mt.factorial(r)*mt.factorial(n-r))
print("Existen ", Cn, "posibles combinaciones sin repetiion del total de ", n , "nombres", "en grupos de ", r, "en", r)
print("Aqui las combinaciones")
combinaciones_reales= list (combinations(nombres,r))

#imprime las combinaciones
for combinations in combinaciones_reales:
  print(combinations)
```
