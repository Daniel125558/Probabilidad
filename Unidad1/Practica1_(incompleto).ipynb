```python
 #definimos una variable llenada tiempos 100mts de tipo losta, que contiene los
 #tiempos que tardaron los 10 alumnos para correr los 100 metros
tiempos100mts= {10,15,17,20,15.25,45,12,14.25,12.36,69.21}
#desplegamos los vaalores de la lista
print(tiempos100mts)

```

```python
#ejemplo 3: histrogramas de 2 grupos
#1. en este ejemplo generamos 100 numeros aleatorios reales que representa la produccion diaria de 100 telares (medidas en metros)
#2. generar otra losta con la produccion de 100 telares para una segunda fabrica

import random
fabrica1=[]
for i in range (100):
  numero_aleatorio= round( random.uniform(10,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2=[]
for i in range (100):
  numero_aleatorio= round( random.uniform(10,100),2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
#libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("produccion diarai de la fabrica1")
```

```python
plt.hist(fabrica2)
plt.title("produccion diaria de la fabrica2")
```

```python
media1=mean(fabrica1)
mediana1=median(fabrica1)
moda1=mode(fabrica1)

print("valores de la fabrica 1 ")
print("la media es: ", media1)
print("la mediana es: ", mediana1)
print("la moda es: ", moda1)
```

```python
media2=mean(fabrica2)
mediana2=median(fabrica2)
moda2=mode(fabrica2)

print("valores de la fabrica 2 ")
print("la media es: ", media2)
print("la mediana es: ", mediana2)
print("la moda es: ", moda2)
```

```python
fig, (ax1,ax2)=plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```
