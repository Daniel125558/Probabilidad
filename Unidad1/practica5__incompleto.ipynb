```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv(data_url)

# The 'horsepower' column in auto-mpg.csv often contains '?'
# Converting '?' to NaN and then to numeric
df['horsepower'] = df['horsepower'].replace('?', np.nan).astype(float)

print("primeras filas del dataset: \n", df.head())

# Use 'weight' column as 'price' is not available and price_column_proxy suggests 'weight'
precios= df["weight"].dropna()
print("\n-------Estadistica basicas de pesos--------")
print("Valor mínimo:", np.min(precios))
print("Valor máximo:", np.max(precios))
print("Moda", precios.mode()[0])
print("Media:", precios.mean())
print("Mediana:", precios.median())
print("Desviación estándar:", precios.std())
```

```python
print("\n------Estadisticos con describe()--------")
print(precios.describe())
```

```python
n=len(precios)
bins=int(1+3.322*math.log10(n))
print("\nCantidad óptima de bins segun Sturges:",bins)
```

```python
plt.figure()
plt.hist(precios, bins=bins)
plt.title("Histograma de Precios de los Automóviles")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```
